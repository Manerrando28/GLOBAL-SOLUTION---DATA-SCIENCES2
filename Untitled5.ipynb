# Preparação dos dados para a Regressão Linear

```python
# Selecionando variáveis categóricas e numéricas
df_reg = df_clean.copy()

# Variáveis independentes (X)
X = df_reg.drop(['salary_in_usd'], axis=1)

# Variável alvo (y)
y = df_reg['salary_in_usd']

# One-Hot Encoding para transformar categorias em números
X_encoded = pd.get_dummies(X, drop_first=True)

# Dividindo em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Escalonamento (normalização)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Dados preparados com sucesso!")
X_encoded.head()
```

### Resposta Dissertativa – Preparação dos Dados para Regressão Linear

Nesta etapa, realizei toda a preparação necessária para treinar o modelo de regressão linear. Primeiro, separei o salário em dólares (`salary_in_usd`) como variável alvo (y) e mantive todas as demais colunas como variáveis independentes (X). Depois disso, apliquei o One-Hot Encoding para transformar todas as categorias, como cargos, modelos de trabalho e localizações, em variáveis numéricas. Esse processo aumentou bastante o número de colunas, chegando a mais de 300 atributos, o que é comum quando trabalhamos com muitos tipos de cargos.

Em seguida, dividi os dados em treino e teste, garantindo que o modelo fosse avaliado de forma justa. Também apliquei a normalização, que é importante para evitar que variáveis com valores maiores influenciem mais do que deveriam durante o treinamento.

No final, o dataset ficou completamente preparado para a etapa de modelagem. Essa preparação garante que o algoritmo consiga lidar corretamente tanto com os dados numéricos quanto com as categorias transformadas.


# Criar o Modelo de Regressão Linear

```python
from sklearn.linear_model import LinearRegression

# Criando o modelo
modelo_lr = LinearRegression()

# Treinando o modelo
modelo_lr.fit(X_train_scaled, y_train)

print("Modelo de Regressão Linear treinado com sucesso!")
```

# Avaliação da Regressão Linear (R², MAE, MSE, RMSE)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Previsões no conjunto de teste
y_pred = modelo_lr.predict(X_test_scaled)

# Métricas
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"R²: {r2:.4f}")
print(f"MAE: {mae:.2f} USD")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f} USD")
```

### Resposta Dissertativa – Avaliação do Modelo de Regressão Linear

A regressão linear apresentou um valor de **R² igual a 0.3565**, o que significa que o modelo conseguiu explicar cerca de 35% da variação dos salários presentes no dataset. Esse valor indica que, embora o modelo consiga capturar parte da relação entre os atributos e o salário, ainda existe uma grande parcela dos fatores que ele não consegue prever — o que é esperado, já que o salário depende de muitas variáveis difíceis de quantificar, como habilidades específicas, desempenho individual e políticas internas das empresas.

O erro médio absoluto (**MAE**) ficou em aproximadamente **39 mil dólares**, mostrando o quanto, em média, as previsões se afastam dos valores reais. Já o **RMSE**, em torno de **61 mil dólares**, reforça que ainda há uma grande variação entre os salários previstos e os observados, especialmente por conta de outliers muito altos.

De forma geral, o modelo funciona como uma boa aproximação inicial, mas não é altamente preciso. Isso faz sentido, já que o dataset possui muitas categorias diferentes (como cargos e localizações) e uma enorme variação salarial, tornando o problema mais complexo para uma regressão linear simples.


# Criando variável binária para Regressão Logística

```python
# Criando a variável binária: 1 = salário acima da mediana, 0 = abaixo
mediana_salario = df_clean['salary_in_usd'].median()
df_clean['salary_high'] = (df_clean['salary_in_usd'] > mediana_salario).astype(int)

# Preparando X e y para a regressão logística
X_log = df_clean.drop(['salary_in_usd', 'salary_high'], axis=1)
y_log = df_clean['salary_high']

# One-hot encoding
X_log_encoded = pd.get_dummies(X_log, drop_first=True)

# Divisão entre treino e teste
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log_encoded, y_log, test_size=0.2, random_state=42
)

# Normalização
scaler_log = StandardScaler()
X_train_log_scaled = scaler_log.fit_transform(X_train_log)
X_test_log_scaled = scaler_log.transform(X_test_log)

print("Dados preparados para Regressão Logística!")
df_clean[['salary_in_usd', 'salary_high']].head()
```

### Resposta Dissertativa – Preparação dos Dados para Regressão Logística

Para aplicar a regressão logística, criei uma variável binária chamada `salary_high`. Ela indica se o salário do profissional está acima da mediana do dataset. Assim, valores maiores que a mediana recebem o rótulo **1**, enquanto salários iguais ou menores recebem **0**. Esse tipo de transformação é comum em problemas de classificação, porque permite treinar um modelo capaz de identificar perfis de profissionais que tendem a ganhar salários mais altos.

Depois de criar essa variável, separei o conjunto de dados em variáveis preditoras (X) e a variável alvo (y). Em seguida, apliquei novamente o One-Hot Encoding para converter todas as categorias em variáveis numéricas e dividi o dataset entre treino e teste. Também foi feita a normalização para garantir que todas as variáveis estivessem na mesma escala, evitando influência indevida de atributos com valores mais altos.



# Treinar a Regressão Logística

```python
from sklearn.linear_model import LogisticRegression

# Criando o modelo
modelo_log = LogisticRegression(max_iter=500)

# Treinando o modelo
modelo_log.fit(X_train_log_scaled, y_train_log)

print("Regressão Logística treinada com sucesso!")
```

# Avaliação da Regressão Logística

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Previsões
y_pred_log = modelo_log.predict(X_test_log_scaled)

# Métricas
acc = accuracy_score(y_test_log, y_pred_log)
prec = precision_score(y_test_log, y_pred_log)
rec = recall_score(y_test_log, y_pred_log)
f1 = f1_score(y_test_log, y_pred_log)

# Matriz de confusão
matriz = confusion_matrix(y_test_log, y_pred_log)

print(f"Acurácia: {acc:.4f}")
print(f"Precisão: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"F1-Score: {f1:.4f}")
print("\nMatriz de Confusão:")
print(matriz)
```

### Resposta Dissertativa – Avaliação da Regressão Logística

A regressão logística apresentou resultados muito satisfatórios para o problema de classificação de salários altos. A acurácia do modelo foi de **91,67%**, indicando que ele acerta a maioria das previsões tanto para salários acima quanto para salários abaixo da mediana.

A precisão ficou em **0.8866**, mostrando que, entre todas as previsões de salários altos feitas pelo modelo, cerca de 88% estavam realmente corretas. Já o recall teve um valor ainda mais alto, **0.9480**, o que significa que o modelo conseguiu identificar quase todos os casos reais de salários altos — um ótimo desempenho para esse tipo de tarefa.

O F1-Score ficou em **0.9163**, confirmando que o modelo mantém um bom equilíbrio entre precisão e recall. Isso é importante porque garante que ele não apenas identifica bem os “altos salários”, mas também evita cometer muitos erros ao fazer essas previsões.

A matriz de confusão mostra que o modelo acertou **608** casos de salários baixos e **621** casos de salários altos. Os erros foram relativamente pequenos: apenas **77 falsos positivos** (salários previstos como altos, mas que não eram) e **33 falsos negativos** (salários altos que o modelo não detectou).

No geral, a regressão logística teve um desempenho excelente, conseguindo diferenciar muito bem os profissionais com salários acima da mediana e mostrando que as variáveis escolhidas possuem um bom poder de previsão nesse tipo de classificação.
